```python
import numpy as np
import pandas as pd
import json
```

## Create a dataframe from the results.json file.

```python
with open("../../results/rich_results.json") as f:
    results_dict = json.load(f)

names, results = zip(*results_dict.items())
df = pd.json_normalize(results)
df.insert(0, "name", names)

df
```

## If the model name is structured as in the example above (which happens automatically from our helper function), we can parse the model, dataset and phase fields.

```python
df.insert(1, "model", df["name"].str.split("_", n=1, expand=True)[0])
df.insert(
    2,
    "dataset",
    df["name"]
    .str.split("_", n=1, expand=True)[1]
    .str.replace("_train$|_test$", "", regex=True),
)
df.insert(3, "phase", df["name"].str.rpartition("_", expand=True)[2])

df
```

## The pandas filter function makes it easily to isoloate on fields of interest ....

```python
df.filter(like="f1")
```

```python
df.filter(like="mcc")
```

```python
df.filter(["model", "dataset", "mcc", "phase", "results.accuracy.f1-score"])
```

## Here is a simple example where we look for the test run with the highest mcc score.

```python
test_df = df[df.phase == "test"]
test_df.loc[test_df["results.accuracy.f1-score"].idxmax()]
```

## We can split the dataframe into train and test data frames and create a merged dataframe where each row represents one complete experiment, i.e., we combine the train data and the test data.

```python
train_df = df[df.phase == "train"]
test_df = df[df.phase == "test"]
experiment_df = pd.merge(
    train_df,
    test_df,
    suffixes=['_train', '_test'],
    how="left",
    left_on=["model", "dataset"],
    right_on=["model", "dataset"],
)
```

## Let's look at the difference in mcc between test and train. . .

```python
df=experiment_df.filter(regex="model|dataset|results.accuracy.mcc").copy()
df['mcc_difference'] = df['results.accuracy.mcc_train'] - df['results.accuracy.mcc_test']
df.sort_values(by=['mcc_difference'])[0:25]
```
