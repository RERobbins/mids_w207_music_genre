## We can split the dataframe into train and test data frames and create a merged dataframe where each row represents one complete experiment, i.e., we combine the train data and the test data.

```python
train_df = df[df.phase == "train"]
test_df = df[df.phase == "test"]
experiment_df = pd.merge(
    train_df,
    test_df,
    suffixes=["_train", "_test"],
    how="left",
    left_on=["model", "dataset"],
    right_on=["model", "dataset"],
)
```

## Let's look at the difference in mcc between test and train. . .

```python
df = experiment_df.filter(regex="model|dataset|result.accuracy.mcc").copy()
df["mcc_difference"] = (
    df["result.accuracy.mcc_train"] - df["result.accuracy.mcc_test"]
)
df.sort_values(by=["mcc_difference"])[0:25]
```
