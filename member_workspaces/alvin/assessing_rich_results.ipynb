```python
import sys
import pandas as pd

from IPython.display import display

sys.path.append("../../")

from helpers.save import results
from helpers.constants import BASE_GENRES
```

```python
save = results('new_rich_results.json')
df = save.unnestDataframeLevel(level = 2)
```

Get top 5 models with highest f1 score per genre in test

```python
for genre in BASE_GENRES:
    columnName = f"result.{genre}.f1-score"

    test = df[df['phase'] == 'test']

    top5 = test.nlargest(5, columnName)
    top5model = top5[['model', 'dataset', 'phase', columnName]].rename(columns={columnName: 'f1_score'})

    print(f'\nFor genre {genre}\n')
    display(top5model)
```
